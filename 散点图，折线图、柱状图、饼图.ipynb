### 散点图

散点图（Scatter Diagram）又称为散点分布图，是利用坐标点（散点）的分布形态反映特征间的统计关系的一种图形。<p>
* 一个特征为横坐标，以另一个特征为纵坐标，值由点在图形中的位置表示。<p>
* 如果有三个特征，且其中一个牲征为类别型特征，散点图可改变该特征的点的形状或颜色，即可了解两个数值型特征和这个类别型特征之间的关系。<p>
---
散点图可以提供两类关键信息，具体内容如下。<p>
* 特征之间是否存在数值或数量的关联趋势，关联趋势是线性的还是非线性的。<p>
* 如果某一个点或某几个点偏离大多数点，那么这些点就是离群值，通过散点图可以一目了然，从而可以进一步分析这些离群值是否在建模分析中产生较大的影响。


matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)<p>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 散点图，s参数设置
np.random.seed(0)
x=np.random.rand(20)
y=np.random.rand(20)
area=(50*np.random.rand(20))**2
 
plt.scatter(x,y,s=area,alpha=0.5)     #气泡图，通过s参数设置点的大小，
plt.show()
```

```python
# 散点图，c参数设置点的颜色
np.random.seed(0)
x=np.random.rand(20)
y=np.random.rand(20)
 
colors=np.random.rand(20)
area=(50*np.random.rand(20))**2
 
plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```

```python
# 散点图，参数marker设置点的形状（P121，表6-7）
np.random.seed(0)
x=np.random.rand(20)
y=np.random.rand(20)
 
colors=np.random.rand(20)
area=(50*np.random.rand(20))**2
 
plt.scatter(x,y,s=area,c=colors,alpha=0.5,marker='x')
plt.show()
```

```python
# 散点图，linewidths设置线条（边线）宽度，只有marker为封闭时才有效。
np.random.seed(0)
x=np.random.rand(20)
y=np.random.rand(20)
 
colors=np.random.rand(20)
area=(50*np.random.rand(20))**2
 
plt.scatter(x,y,s=area,c=colors,alpha=0.5,marker='D',linewidths=5)
plt.show()
```

```python
# 例6-20  （参数edgecolors,边缘颜色)
fig,ax=plt.subplots()
for color in ['red','green','blue']:
    n=300
    x,y=np.random.randn(2,n)   #x，y为正态分布的n个数
    ax.scatter(x,y,c=color,label=color,alpha=0.3,edgecolors='none') #edgecolors边缘颜色
    ax.legend()
    ax.grid(True)

plt.show()
```

### 折线图

折线图（Line Chart）是一种将数据点按照顺序连接起来的图形，可以看作是将散点图按照x轴坐标顺序连接起来的图形。<p>
主要功能：<p>
* 查看因变量y随着自变量x改变的趋势，适合用于显示随时间（根据常用比例设置)而变化的连续数据。<p>
* 显示数量的差异和增长趋势的变化。<p>
---


```python
#折线图 
#plot函数一次可以接收多组数据，添加多条折线图，同时还可以分别定义每条折线的颜色、点的形状和类型，将折线的颜色、点的形状和类型连接在一起用一个字符串表示
import numpy as np
x = np.random.randint(0,20,5)
y1 = 2*x
y2 = 2*x*x + 1

p = plt.figure(figsize=(6,6))

plt.plot(x,y1,color='r',linestyle='-.',marker='*',alpha=0.7)
# plt.plot(x,y1,'r-.x')                              #等同于上一句

#同时绘制多条折线
# plt.plot(x,y1,'r-.x',x,y2,'g*:')
# plt.legend(['$y=2x$','$y=2x^2+1$'])         #在$$之间进行公式编辑，P126

plt.show()
```

### 柱形图

柱形图（Bar Chart）的核心思想是对比，因为人眼对柱形高度差异很敏感，辨识效果非常好，它常用于显示一段时间内的数据变化或显示各项数据之间的比较情况。<p>
适用场合：二维数据集（每个数据点包括两个值x和y），但只有一个维度的值需要比较。<p>
局限：只适用于中小规模的数据集。
---


```python
# 柱形图，基础参数设置
plt.figure(figsize=(13,9))

plt.subplot(231)
x=['c','b','a']
hight=[1,3,2]
a=plt.bar(x,hight)                # 按默认值绘制
plt.title('默认样式')

plt.subplot(232)
x=['a','b','c']
hight=3
width=0.2
plt.bar(x,hight,width)                   #width参数用于设置宽度    
plt.title('设置同一高度、自定义宽度')

plt.subplot(233)
x=['a','b','c']
width=[0.3,0.2,0.4]
hight=[1,3,2]
a=plt.bar(x,hight,width)                 #                
plt.title('自定义宽度')

plt.subplot(234)
x=['a','b','c']
hight=[1,3,2]
width=0.2
bottom=0.2
plt.ylim(0,4)
plt.bar(x,hight,width,bottom)       #bottom参数用于设置基准位置
plt.title('自定义基准位置')

plt.subplot(235)
x=['a','b','c']
hight=[1,3,2]
plt.bar(x,hight,align='edge')      #align参数可选center（默认）和edge左侧对齐。
plt.title('设置左侧对齐')

plt.subplot(236)
x=['a','b','c']
hight=[1,3,2]
width=-0.8
plt.bar(x,hight,width,align='edge')    #设置右侧对齐时需同时设置width为负值，且align='edge'。
plt.title('设置右侧对齐')

plt.show()
```

```python
#柱形图，外观参数
plt.figure(figsize=(13,9))

plt.subplot(231)
x=['a','b','c']
hight=[1,3,2]
plt.bar(x,hight)
plt.title('默认样式')

plt.subplot(232)
x=['a','b','c']
hight=[1,3,2]
plt.bar(x,hight,color='r',edgecolor='g',linewidth=2)    
plt.title('自定义柱子颜色、边框颜色、线宽')

plt.subplot(233)
x=['a','b','c']
hight=[1,3,2]
plt.bar(x,hight,tick_label=[1,2,3])      #label为设置图表标签，用于显示图例
plt.title('设置刻度标签')

plt.subplot(234)
x=['a','b','c']
hight=[1,3,2]
plt.bar(x,hight,hatch='/')
plt.title('设置柱子填充符号')

plt.subplot(235)
x=['a','b','c']
hight=[1,3,2]
plt.bar(x,hight,hatch='//')
plt.title('重复填充符号')

plt.subplot(236)
x=['a','b','c']
hight=[1,3,2]
plt.bar(x,hight,hatch='/+')
plt.title('组合填充符号')

plt.show()
```

```python
# 简单柱状图,显示数据文本
p = plt.figure(figsize=(8,6))

xiaoming_score = [80,75,65,58,75,80,90]  #小明各科成绩
xiaohong_score =[90,85,75,62,75,60,80]   #小红各科成绩
subjects = ['语文','英语','数学','物理','化学','生物','体育']


plt.bar(subjects,xiaohong_score)
# plt.bar(subjects,xiaoming_score)     #因为都是按照默认值绘图，所以两张图柱形重叠。

# 显示数据文本
for i in range(len(subjects)):
    plt.text(i,xiaohong_score[i]+1,xiaohong_score[i])

plt.show()
```

```python
#簇状柱形图
p = plt.figure(figsize=(8,6))

plt.bar(subjects,xiaohong_score,width=-0.4,align='edge',label='小红')
plt.bar(subjects,xiaoming_score,width=0.4,align='edge',label='小明') 
plt.title('小明、小红各科成绩对比') 
plt.xlabel('科目',fontsize = 15)   
plt.ylabel('成绩',fontsize = 15) 
plt.legend(loc = 'upper center')
plt.show()
```

```python
#堆叠柱状图
p = plt.figure(figsize=(8,6))
plt.bar(subjects,xiaohong_score,label='小红')
plt.bar(subjects,xiaoming_score,bottom=xiaohong_score,label='小明') #用bottom参数实现堆叠效果

plt.legend()

plt.show()
```

```python
# 条形图
plt.barh(subjects,xiaohong_score)
plt.show()
```

### 饼图

饼图（Pie Graph）是将各项的大小与各项总和的比例显示在一张“饼”中，以“饼”的大小来确定每一项的占比。<p>
饼图可以清晰的反映出部分与部分，部分与整体之间的比例关系，易于显示每组数据相对于总数的比例，而显示方式直观。<p>
---
matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=0, 0, frame=False, rotatelabels=False, *, normalize=None, data=None)

```python
#饼图，用explode（偏移距离）参数可设置突出显示的块,startangle用于设置饼图起始摆放角度，即”第一块的起始位置“

labels = ['Frogs', 'Hogs', 'Dogs', 'Logs'] # 定义标签
sizes = [15, 30, 45, 10]  # 每一块的比例
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']  # 每一块的颜色
explode = (0, 0.1, 0, 0)  # 突出显示，这里仅仅突出显示第二块（即'Hogs'）

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90) #shadow设置阴影，startangle从经轴逆时针旋设置饼的角度。
plt.axis('equal')  # 显示为圆（避免比例压缩为椭圆）
plt.show()
```

```python
# 饼图，pctdistance默认为0.6,labeldistance默认为1.1，当它们值大于1时，显示在圆外
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',pctdistance=1.3,labeldistance=0.5)      #设置标签显示圆内，数据显示圆外

plt.show()
```

```python
#环形图：通过参数wedgeprops可设置饼图内外边界的属性，如边界线的粗细、颜色等
x=[2,8,45,15]

plt.pie(x,autopct='%1.1f%%',pctdistance=0.8,wedgeprops = {'width':0.5})

plt.show()
```

```python
# 内嵌环形图，适用于作比例对比。绘制时注意内、外环要使用同样的颜色序列，通过设置内、外环不同的半径radius达到效果。
x1=[2,8,45,15]
x2=[5,21,32,13]
c = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'] 
plt.pie(x1,colors=c,autopct='%1.1f%%',pctdistance=0.8,wedgeprops = {'width':0.3})
plt.pie(x2,colors=c,autopct='%1.1f%%',radius=0.7,pctdistance=0.8,wedgeprops = {'width':0.3})   #注意半径的0.7+上面外环宽度刚好为1（外环半径）
plt.show()
```
