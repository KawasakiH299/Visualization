### 二 pyecharts绘制直角坐标系图表

直角坐标系图表包含：<p>
Bar：柱状图<p>
Boxplot:箱形图<p>
EffectScatter：涟漪特效散点图<p>
HeatMap：热力图<p>
Kline/Candlestick：k线图<p>
Line：折线图/面积图<p>
Scatter：散点图<p>
Overlap:层叠多图<p>

直角坐标系图表继承自 RectChart 都拥有以下方法:<p>
extend_axis  # 扩展 X/Y 轴<p>
add_xaxis  # 新增 X 轴数据<p>
reversal_axis  # 翻转 XY 轴数据<p>
overlap  # 层叠多图<p>

例：Bar:柱状图/条形图<p>
柱状图/条形图(Bar) ，使用垂直的柱子显示类别之间的数值比较。其中一个轴表示需要对比的分类维度，另一个轴代表相应的数值。<p>
对比分类数据的数值大小，当分类情况过多时，柱状图的文本为了排布合理，需要进行旋转，不利于阅读，相比于纵向柱状图，横向柱状图更适用于此类分类较多的场景。<p>


```python

from pyecharts.faker import Faker    #https://blog.csdn.net/weixin_49432344/article/details/123872327
from pyecharts.charts import Bar
print(Faker.phones)
print(Faker.values())     #随机生成
c = Bar()
c.add_xaxis(Faker.phones)
c.add_yaxis('A',Faker.values())
c.add_yaxis('B',Faker.values())
c.render_notebook()
```

```python
#例8-3
from pyecharts.charts import Bar
from pyecharts import options as opts
%matplotlib inline
bar = Bar()
bar.add_xaxis(["衬衫", "毛衣", "领带", "裤子", "风衣", "高跟鞋", "袜子"])
bar.add_yaxis("商家A", [114, 55, 27, 101, 125, 27, 105])
bar.add_yaxis("商家B", [57, 134, 137, 129, 145, 60, 49])
# 设置显示标题，副标题和工具箱
bar.set_global_opts(title_opts=opts.TitleOpts(title="货品销售情况",subtitle = "A和B公司"),toolbox_opts = opts.ToolboxOpts(is_show = True))
bar.set_series_opts(label_opts=opts.LabelOpts(position = "right"))    #数据标签位置
bar.reversal_axis()     #翻转XY轴
bar.render_notebook()
```

绘制特效散点图type_=ChartType.EFFECT_SCATTER

```python
from pyecharts.charts import Geo
import pyecharts.options as opts
from pyecharts.globals import ChartType      #
from pyecharts.faker import Faker
# from example.commons import Faker

(
    Geo()
    .add_schema(maptype='china')
    .add(series_name='', data_pair=[(i, j) for i, j in zip(Faker.provinces, Faker.values())],
        type_=ChartType.EFFECT_SCATTER)     
    .set_global_opts(
        title_opts=opts.TitleOpts(title='中国地图(特效散点图)'),
        visualmap_opts=opts.VisualMapOpts(
            is_piecewise=True
        )
    )
).render_notebook()
```

```python
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType
import pyecharts.options as opts

city_num = [('广州', 105), ('成都', 70), ('北京', 99), ('西安', 80)]
start_end = [('广州', '成都'), ('广州', '北京'), ('广州', '西安')]


(
    Geo()
    .add_schema(maptype='china', 
                itemstyle_opts=opts.ItemStyleOpts(color='#323c48', border_color='#111'))
    .add('', data_pair=city_num, color='white')
    .add('', data_pair=start_end, type_=ChartType.LINES,
         effect_opts=opts.EffectOpts(symbol=SymbolType.ARROW, 
                                     color='blue', 
                                     symbol_size=8))
).render_notebook()
```

```python
#例8-18
from pyecharts import options as opts 
from pyecharts.charts import Geo 
from pyecharts.globals import ChartType, SymbolType

c = (
        Geo()
        .add_schema(maptype="china")
        .add( "",
            [ ("哈尔滨", 66), ("重庆", 88), ("上海", 100), ("乌鲁木齐", 30),("北京", 30),("兰州",170)],
            type_=ChartType.EFFECT_SCATTER,           #设置点的涟漪效果
            color="green",
        )
        .add(
            "geo",
            [("北京", "兰州"),( "兰州","北京"), ("重庆", "杭州"),("哈尔滨", "重庆"),("乌鲁木齐", "哈尔滨")],
            type_=ChartType.LINES,                  #设置点间连线
            effect_opts=opts.EffectOpts(
                symbol=SymbolType.ARROW, symbol_size=6, color="blue"
            ),
            linestyle_opts=opts.LineStyleOpts(curve=0.2),
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title="主要城市航班路线和数量"))
    )

c.render_notebook()
```
